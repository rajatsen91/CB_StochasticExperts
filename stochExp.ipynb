__Processing Stream DataSet into required format__

```python
raw_data = pd.read_table('./data/stream_labeled_data.vw',header=None)
data_series = raw_data[0]
K = 10

def process_string(s):
    s = s.split()
    label = int(s[0])
    a = []
    for i in range(2,len(s)):
        a = a + [int(s[i].split(':')[1])]

    a = a + [label]
    return np.array(a)

collection = data_series.apply(process_string)
matrix = collection.as_matrix()
data = np.vstack(matrix)
Xdata = data[:,0:100].astype(float) #feature/context matrix created
Ydata = data[:,100]

Rdata = np.zeros([len(Ydata),K])
for i in range(len(Ydata)):
    Rdata[i,Ydata[i]-1] = 1.0   #reward matrix created
```

__Initializing Base class with default parameters for running MoM algorithm__

```python
SE = stochExp(Xdata = Xdata,Rdata = Rdata,K = 10,T0 = 50,C1 = 2.0 ,C2 = 1.0,isMoM = True,calibrate = False,bsize_mult = 7,initial_model = [3,3,1], model_ratio = [3,1],max_depths = [3,6,10], n_estimators = [50,100,200], colsample_bytrees = [0.6,0.8],  nthread = 16, penalty = ['l1','l2'], C = [1e-3,1e-2,1e-1,1.0,1e1,1e2],log_file = './examples/results/rewards_stream_MoM.npy',cl_limit = 1.0)
```

__Running MoM Algorithm__

```python
SE.run_algorithm()  #may take time for larger data-sets
```
